# Introduction to The Cancer Genome Atlas (TCGA)

## What is TCGA?

**The Cancer Genome Atlas (TCGA)** is a landmark dataset developed by the National Cancer Institute (NCI) and the National Human Genome Research Institute (NHGRI). It provides an expansive resource of genomic, transcriptomic, and clinical data across multiple cancer types, enabling researchers to delve into the molecular underpinnings of cancer. TCGA has revolutionized cancer genomics by enabling large-scale integrative analyses and fostering advancements in precision oncology.

---

## Why Focus on Colon Adenocarcinoma (TCGA-COAD) and Rectum Adenocarcinoma (TCGA-READ)?

For this study, I selected two colorectal cancer subtypes from TCGA:
1. **Colon Adenocarcinoma (Project ID: TCGA-COAD)**  
2. **Rectum Adenocarcinoma (Project ID: TCGA-READ)**

These cancers collectively represent **colorectal cancer**, the third most common cancer globally. While both arise in the gastrointestinal tract, their differing anatomical locations and subtle molecular distinctions warrant separate analyses. Integrating data from both cancers allows us to explore shared and unique features of these subtypes.

### Overview of Colon Adenocarcinoma (COAD):
- **Origin**: The large intestine (colon).
- **Characteristics**: Often associated with genetic mutations in key pathways like **Wnt signaling** (e.g., **APC**, **KRAS**) and **TP53**.
- **Clinical Features**: Symptoms include changes in bowel habits, abdominal discomfort, and rectal bleeding. The disease may be influenced by dietary and lifestyle factors.

### Overview of Rectum Adenocarcinoma (READ):
- **Origin**: The rectum, the final segment of the gastrointestinal tract.
- **Characteristics**: Similar molecular features to COAD but may differ in mutation frequency and treatment response.
- **Clinical Features**: Rectal bleeding, tenesmus, and altered bowel movements are common symptoms.

---

## TCGA Data and Its Utility in This Analysis

TCGA provides a comprehensive dataset for **COAD** and **READ**, enabling integrative analyses of their similarities and differences. Key data used for this study include:
1. **Gene Expression Profiles** (e.g., mRNA expression): To identify key transcriptional changes.

# Supervised Learning Pipeline Description

## Data Overview
The dataset used in this analysis includes the following key variables:
- **SurvivalTime**: The time (in units) until an event (e.g., death or relapse) or censoring occurred.
- **SurvivalStatus**: A binary variable indicating whether the event occurred (1) or the data was censored (0).

This survival data forms the foundation for assessing the relationship between gene expression and survival outcomes.

---

## Gene Selection: Differential Gene Expression Analysis
Before proceeding with the survival analysis, a differential gene expression analysis was performed to identify genes of interest. Specifically:

- The top 50 up-regulated genes (based on log2 fold change and adjusted p-value) were selected from the analysis.
- Similarly, the top 50 down-regulated genes were chosen.

This resulted in a total of **100 differentially expressed genes** being identified as the initial set of genes of interest for further analysis. These selected genes represent those most significantly associated with changes in expression levels under specific experimental or clinical conditions. By focusing on these genes, the analysis aims to explore their potential impact on survival outcomes.

---

## Survival Analysis Pipeline

### Preprocessing and Cleaning Data:
- Expression data corresponding to the 100 selected genes was extracted.
- Missing or infinite values were replaced with column means to ensure data consistency.

### Feature Selection Using Univariate Cox Regression:
- Each of the 100 genes was independently tested for its association with survival outcomes using a **univariate Cox Proportional Hazards model**.
- Genes with a p-value < 0.05 were deemed significant and retained for further modeling.

### Survival Modeling:
The significant genes were used as features to build and evaluate three survival models:
1. **Cox Proportional Hazards Model (CoxPH)**.
2. **Penalized Cox Model** to handle high-dimensional data.
3. **Random Survival Forest (RSF)** for capturing non-linear relationships.

### Evaluation and Interpretation:
- Risk scores, concordance indices, and ROC curves were used to compare model performances.
- Key genes influencing survival were identified, and their biological significance was explored.

---

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from lifelines import CoxPHFitter
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sksurv.util import Surv
from sksurv.metrics import concordance_index_censored, cumulative_dynamic_auc
from sksurv.linear_model import CoxPHSurvivalAnalysis, CoxnetSurvivalAnalysis
from sksurv.ensemble import RandomSurvivalForest
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score

# Function to compute dynamic AUC over multiple time points
def compute_dynamic_auc(y_train_structured, y_test_structured, risk_scores, time_points):
    """
    Compute dynamic AUC over multiple time points.

    Parameters:
    - y_train_structured: Structured array of training survival data.
    - y_test_structured: Structured array of test survival data.
    - risk_scores: Array-like, predicted risk scores from the model.
    - time_points: Array-like, time points at which to compute AUC.

    Returns:
    - aucs: Array of AUC values corresponding to the time_points.
    """
    # Compute cumulative dynamic AUC
    aucs, _ = cumulative_dynamic_auc(
        y_train_structured,
        y_test_structured,
        risk_scores,
        time_points
    )
    return aucs

# Load the merged dataset
merged_data = pd.read_csv('merged_dataset.csv')

# Clean column names by stripping whitespace and removing quotes
merged_data.columns = merged_data.columns.str.strip().str.replace('"', '').str.replace("'", '')

# Load the gene list from CSV
genelist_df = pd.read_csv('gene_list.csv')
genelist = genelist_df['Gene'].tolist()

# Convert 'SurvivalTime' and 'SurvivalStatus' to numeric, coercing errors to NaN
merged_data['SurvivalTime'] = pd.to_numeric(merged_data['SurvivalTime'], errors='coerce')
merged_data['SurvivalStatus'] = pd.to_numeric(merged_data['SurvivalStatus'], errors='coerce')

# Drop rows with missing survival data
merged_data = merged_data.dropna(subset=['SurvivalTime', 'SurvivalStatus'])

# Identify gene columns present in the data
gene_columns = [gene for gene in genelist if gene in merged_data.columns]

# Inform about genes not found in the data
genes_not_found = [gene for gene in genelist if gene not in merged_data.columns]
if genes_not_found:
    print(f"The following genes were not found in the data and will be ignored: {genes_not_found}")

# Replace infinite values with NaN and fill missing gene expression values with the mean
merged_data[gene_columns] = merged_data[gene_columns].replace([np.inf, -np.inf], np.nan)
merged_data[gene_columns] = merged_data[gene_columns].fillna(merged_data[gene_columns].mean())

# Drop any remaining rows with NaNs in gene expression data
merged_data = merged_data.dropna(subset=gene_columns)

# Define features (X) and target variables (y)
X = merged_data[gene_columns]
y = merged_data[['SurvivalTime', 'SurvivalStatus']]

# Ensure no NaNs remain in features and target
print("Checking for NaNs in features and target variables...")
print(f"Total NaNs in X: {X.isnull().sum().sum()}")
print(f"Total NaNs in y: {y.isnull().sum().sum()}")

# Split the data into training and validation sets (70% train, 30% validate)
X_train, X_validate, y_train, y_validate = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=123,
    stratify=y['SurvivalStatus']
)

print(f"Training set size: {X_train.shape[0]} samples")
print(f"Validation set size: {X_validate.shape[0]} samples")

# Initialize an empty list to hold significant genes from univariate Cox regression
significant_genes = []

# Perform univariate Cox regression for each gene to filter significant ones
for gene in gene_columns:
    # Combine gene expression with survival data and drop any NaNs
    df = pd.concat([X_train[gene], y_train], axis=1).dropna()
    
    # Initialize CoxPHFitter from lifelines
    cph = CoxPHFitter()
    
    try:
        # Fit the Cox model
        cph.fit(df, duration_col='SurvivalTime', event_col='SurvivalStatus')
        
        # Extract the p-value for the gene
        p_value = cph.summary.loc[gene, 'p']
        
        # If p-value is less than 0.05, consider the gene significant
        if p_value < 0.05:
            significant_genes.append(gene)
    except Exception as e:
        print(f"Error fitting Cox model for gene {gene}: {e}")

print(f"Number of significant genes: {len(significant_genes)}")
print("Significant genes:")
print(significant_genes)

# Check if there are any significant genes
if len(significant_genes) == 0:
    raise ValueError("No significant genes found. Please adjust your criteria or check the data.")

# Subset training data to include only significant genes
X_train_sig = X_train[significant_genes]

# Scale the features using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_sig)
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train_sig.columns, index=X_train_sig.index)

# Prepare structured survival data using sksurv
y_train_structured = Surv.from_dataframe('SurvivalStatus', 'SurvivalTime', y_train)

# Subset validation data to include only significant genes
X_validate_sig = X_validate[significant_genes]

# Scale the validation features using the same scaler fitted on training data
X_validate_scaled = scaler.transform(X_validate_sig)
X_validate_scaled = pd.DataFrame(X_validate_scaled, columns=X_validate_sig.columns, index=X_validate_sig.index)

# Prepare structured survival data for validation
y_validate_structured = Surv.from_dataframe('SurvivalStatus', 'SurvivalTime', y_validate)

# Initialize and fit the Cox Proportional Hazards model using scikit-survival
cox_model = CoxPHSurvivalAnalysis()
cox_model.fit(X_train_scaled, y_train_structured)

# Predict risk scores on the validation set using the Cox model
cox_risk_scores = cox_model.predict(X_validate_scaled)

# Calculate the concordance index for the Cox model on the validation set
c_index_cox = concordance_index_censored(
    y_validate['SurvivalStatus'].astype(bool).values,
    y_validate['SurvivalTime'].values,
    -cox_risk_scores  # Negative sign if higher scores indicate higher risk
)[0]
print(f"CoxPH Model C-index on validation set: {c_index_cox:.4f}")

# Initialize and fit the Random Survival Forest (RSF) model
rsf_model = RandomSurvivalForest(
    n_estimators=100,
    min_samples_leaf=5,
    random_state=123
)
rsf_model.fit(X_train_scaled, y_train_structured)

# Predict risk scores on the validation set using the RSF model
rsf_risk_scores = rsf_model.predict(X_validate_scaled)

# Calculate the concordance index for the RSF model on the validation set
c_index_rsf = concordance_index_censored(
    y_validate['SurvivalStatus'].astype(bool).values,
    y_validate['SurvivalTime'].values,
    rsf_risk_scores
)[0]
print(f"Random Survival Forest Model C-index on validation set: {c_index_rsf:.4f}")

# Initialize and fit the Penalized Cox Proportional Hazards model
coxnet_model = CoxnetSurvivalAnalysis(l1_ratio=0.5, alphas=[0.1, 0.01, 0.001], max_iter=10000, tol=1e-4)
coxnet_model.fit(X_train_scaled, y_train_structured)

# Predict risk scores on the validation set using the Penalized Cox model
coxnet_risk_scores = coxnet_model.predict(X_validate_scaled)

# Calculate the concordance index for the Penalized Cox model on the validation set
c_index_coxnet = concordance_index_censored(
    y_validate['SurvivalStatus'].astype(bool).values,
    y_validate['SurvivalTime'].values,
    -coxnet_risk_scores  # Negative sign if higher scores indicate higher risk
)[0]
print(f"Penalized CoxPH Model C-index on validation set: {c_index_coxnet:.4f}")

# Define multiple time points for AUC calculation (e.g., 100 evenly spaced points)
time_points = np.linspace(
    y_validate['SurvivalTime'].min(), 
    y_validate['SurvivalTime'].max(), 
    100, 
    endpoint=False  # Exclude the maximum time point
)

# Verify that all time points are within the valid range
max_time = y_validate['SurvivalTime'].max()
if np.any(time_points >= max_time):
    raise ValueError("Some time points exceed or match the maximum survival time in the validation set.")
else:
    print("All time points are within the valid range.")

# Compute dynamic AUCs for each model
cox_auc = compute_dynamic_auc(
    y_train_structured, 
    y_validate_structured, 
    -cox_risk_scores, 
    time_points
)
rsf_auc = compute_dynamic_auc(
    y_train_structured, 
    y_validate_structured, 
    rsf_risk_scores, 
    time_points
)
coxnet_auc = compute_dynamic_auc(
    y_train_structured, 
    y_validate_structured, 
    -coxnet_risk_scores, 
    time_points
)

# Plot AUC vs. Time for all models
plt.figure(figsize=(12, 8))
plt.plot(time_points, cox_auc, label=f'CoxPH (AUC)')
plt.plot(time_points, rsf_auc, label=f'Random Survival Forest (AUC)')
plt.plot(time_points, coxnet_auc, label=f'Penalized CoxPH (AUC)')
plt.xlabel('Time')
plt.ylabel('AUC')
plt.title('Dynamic AUC Over Time for Survival Models')
plt.legend()
plt.grid(True)
plt.show()
```

## Consensus Clustering: Identifying Subgroups
To identify subgroups of tumor samples based on immune infiltration patterns, the following steps were performed:

### Preprocessing and Cleaning Data:
- The immune infiltration data was loaded from `im_ssgsea.csv`.
- Rows represent tumor samples, and columns represent immune-related features such as immune cell types or immune pathways.

### Consensus Clustering Algorithm:
- A **Consensus Clustering** approach was implemented using the `ConsensusCluster` class. Key parameters included:
  - **Clustering Range**: Evaluated cluster numbers from **2 to 8** (`L=2, K=9`).
  - **Resampling**: Used 80% of the data in each resampling iteration to assess cluster stability.
  - **Resampling Iterations**: Performed **100 iterations** (`H=100`) for each value of `k`.
- **K-Means Clustering** was used as the base clustering method.

### Optimal Number of Clusters:
- The algorithm determined the best number of clusters (`bestK`) by analyzing the change in clustering stability (`K`) across different values of `k`.

## Results and Analysis of Clustering
### Key Output:
- The optimal number of clusters (`bestK`) was found to be **3**, indicating that the tumor samples can be grouped into **three distinct subtypes** based on their immune infiltration profiles.

### Biological Implications:
1. **Cluster Subtypes:**
   - These clusters likely represent biologically distinct **immune subtypes** of tumors:
     - **Cluster 1**: Tumors with high immune infiltration ("hot" tumors).
     - **Cluster 2**: Tumors with low immune infiltration ("cold" tumors).
     - **Cluster 3**: Tumors with mixed or intermediate immune profiles.
2. **Therapeutic Relevance:**
   - **Hot tumors** may respond better to immunotherapy.

3. **Hypothesis Generation:**

## Conclusion:
The analysis successfully identified **three distinct tumor subtypes** based on immune infiltration profiles using a robust **Consensus Clustering** approach. These findings provide insights into tumor immune microenvironments and their potential implications for therapeutic strategies. Further validation and integration with clinical data are recommended to enhance the biological and clinical relevance of these results.

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from itertools import combinations
import bisect

# Load the im_ssgsea.csv data directly
data = pd.read_csv("im_ssgsea.csv", index_col=0)

# Define the ConsensusCluster class
class ConsensusCluster:
    def __init__(self, cluster, L, K, H, resample_proportion=0.8):
        assert 0 <= resample_proportion <= 1, "proportion must be between 0 and 1"
        self.cluster_ = cluster
        self.L_ = L
        self.K_ = K
        self.H_ = H
        self.resample_proportion_ = resample_proportion
        self.Mk = None
        self.Ak = None
        self.deltaK = None
        self.bestK = None

    def _internal_resample(self, data, proportion):
        resampled_indices = np.sort(np.random.choice(
            range(data.shape[0]), size=int(data.shape[0] * proportion), replace=False))
        return resampled_indices, data.iloc[resampled_indices, :]

    def fit(self, data, verbose=False):
        Mk = np.zeros((self.K_ - self.L_, data.shape[0], data.shape[0]))
        Is = np.zeros((data.shape[0],) * 2)
        for k in range(self.L_, self.K_):
            i_ = k - self.L_
            if verbose:
                print(f"At k = {k}, iteration = {i_}")
            for h in range(self.H_):
                resampled_indices, resample_data = self._internal_resample(data, self.resample_proportion_)
                Mh = self.cluster_(n_clusters=k).fit_predict(resample_data)
                index_mapping = np.array((Mh, resampled_indices)).T
                index_mapping = index_mapping[index_mapping[:, 0].argsort()]
                sorted_ = index_mapping[:, 0]
                id_clusts = index_mapping[:, 1]
                for i in range(k):
                    ia = bisect.bisect_left(sorted_, i)
                    ib = bisect.bisect_right(sorted_, i)
                    is_ = np.sort(id_clusts[ia:ib])
                    ids_ = np.array(list(combinations(is_, 2))).T
                    if ids_.size != 0:
                        Mk[i_, ids_[0], ids_[1]] += 1
                ids_2 = np.array(list(combinations(resampled_indices, 2))).T
                Is[ids_2[0], ids_2[1]] += 1
            Mk[i_] /= Is + 1e-8
            Mk[i_] += Mk[i_].T
            np.fill_diagonal(Mk[i_], 1)
            Is.fill(0)
        self.Mk = Mk
        self.Ak = np.array([np.trapz(np.cumsum(np.histogram(m.ravel(), bins=100, density=True)[0]))
                            for m in Mk])
        self.deltaK = np.diff(self.Ak) / self.Ak[:-1]
        self.bestK = np.argmax(self.deltaK) + self.L_

    def predict(self):
        assert self.Mk is not None, "First run fit"
        return self.cluster_(n_clusters=self.bestK).fit_predict(1 - self.Mk[self.bestK - self.L_])

# Perform Consensus Clustering
consensus_cluster = ConsensusCluster(KMeans, L=2, K=9, H=100, resample_proportion=0.8)
consensus_cluster.fit(data, verbose=True)
print(f"The best K (number of clusters) is: {consensus_cluster.bestK}")
```
